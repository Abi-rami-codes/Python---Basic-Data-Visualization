### 50. Write a Python program to convert a string to datetime.
Sample String : Jul 1 2014 2:43PM 

Expected Output : 2014-07-01 14:43:00

```python
def toDatetime(date):
    d2 = pd.to_datetime(date, format ='%b %d %Y %I:%M%p')
    return d2

toDatetime('Jul 1 2014 2:43PM')
```

### 52. Write a Python program to convert unix timestamp string to readable date.

Sample Unix timestamp string : 1284105682
    
Expected Output : 2010-09-10 13:31:22

```python
def unixtodate(unix):
    time = (datetime.fromtimestamp(int(unix))).strftime('%Y-%m-%d %H:%M:%S')
    return time

unixtodate('1284105682')
```

### 54. Write a Python program to get days between two dates. 

```python
def daydifference(date1, date2):
    date1 = pd.to_datetime(date1, format = "%Y-%m-%d")
    date2 = pd.to_datetime(date2, format = "%Y-%m-%d")
    diff = date2 - date1
    return diff

print(daydifference("2019-6-16", "2018-4-10"))

date3 = datetime.today()
date4 = pd.to_datetime(2019-3-22)

date3 - date4
```
